## **Descripción de la Base de Datos**:

**Variables**:

1. **Mes:** Mes en el que ocurrio la venta.-
2. **Año:** Año en el que ocurrio la venta.-
3. **Grupo:** Indica el grupo al cual pertenece los insumos vendidos.-
4. **Rubro:** Indica el rubro al cual pertenete los insumos vendidos.-
5. **Vendedor:** Indica la persona que ejecuto la venta.-
6. **Comprobante de Venta:** Es el numero de comprobante con el cual se documenta la venta.-
7. **Cantidad:** Indica la cantidad de insumos vendidos.-
8. **Unitario USD:** Indica el precio unitario en USD.-
9. **Total USD:** Indica el total el USD vendidos por insumo.
10. **Unidades:** Indica las unidades del insumo de acuerdo a su forma de envasado.-
11. **Total Unidades:** Indica la cantidad de unidades vendidas.-
12. **Precio Costo:** Es el precio de costo de los insumos.-
13. **Total Costo:** Indica el total de costo por insumos vendidos.-
14. **Diferencia Unitaria:** Es la diferencia por unidad entre el precio de venta y el precio de costo.-
15. **Diferencia Total:** Es la diferencia total entre el precio de costo y el precio de venta.-
16. **Porcentaje de Ganancia:** Es el porcentaje de ganancia que se obtiene por la venta de cada insumo sobre el precio de costo.-

```python
# Vamos a usar esta función para detectar todos los valores atipicos del Boxplot:

def find_outliers_IQR(df,feature):
   q1=df[feature].quantile(0.25)
   q3=df[feature].quantile(0.75)
   IQR=q3-q1
   outliers = df[((df[feature]<(q1-1.5*IQR)) | (df[feature]>(q3+1.5*IQR)))]
   return outliers
```

```python
atipicos = find_outliers_IQR(df,"Porcent_Ganancia")
atipicos
```

## **Normalizacion de Datos y Generacion del Modelo (Kmeans)**:

Normalizamos los datos obtenidos de nuestro sub dataset, para lograr valores mas estandarizados para analizar.- Vamos a optar por un logaritmo de regresion y nuestro elegido va a ser K-Means. Como aprendimos, este algoritmo separa el conjunto de datos dado en  k  grupos que minimizan la suma de los cuadrados de distancias entre cada par de puntos en el grupo.
Se van a ejecutar los codigos correspodientes para preseleccionar el valor de  k , que puede requerir un análisis previo para determinarlo. Y asi obtener el numero de clusters optimo a traves del metodo de codo.- 

```python
def normalize(df_data):
    result = df.copy()
    for feature_name in df.columns:
        max_val = df[feature_name].max()
        min_val = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_val) / (max_val - min_val)
        
    return result
```

```python
df_data.head()
```

```python
df_norm = df_data[['Total_USD','Porcent_Ganancia']]
df_norm.head()
```

```python
wcss = []
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Creamos un gráfico
fig,ax = plt.subplots(figsize =(18, 8))
plt.plot(range(1, 21), wcss,color='Darkgreen')
plt.title('Metodo del codo')
plt.xlabel('Numero de clusters')
plt.ylabel('Inercia')
plt.show()
```

><font color="006400">**A partir del gráfico, parecería que el valor de agregar más grupos realmente comienza a disminuir alrededor de k=3.-**

```python
k = 4 # Definimos los Grupos en base al resultado obtenido con el Grafico de Codo
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=0)
kmeans.fit(df_norm)
```

```python
labels = kmeans.predict(df_norm)
centroids = kmeans.cluster_centers_
centroids
```

## **Métrica para validar el modelo K-Means**:
 **Metodo de Silhouette**: 

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
y_predict = []
for i,k in enumerate([1,2,3,4]):    
    
    km = KMeans(n_clusters=k)
    y_predict = km.fit_predict(df_norm)
    centroids  = km.cluster_centers_

silhouette_vals = silhouette_samples(df_norm,y_predict)
```

```python
np.unique(y_predict)
```

```python
y_upper = 0

for i, cluster in enumerate(np.unique(y_predict)):
  cluster_silhouette_vals = silhouette_vals[y_predict == cluster]
  cluster_silhouette_vals.sort()
  y_upper += len(cluster_silhouette_vals)


  avg_score = np.mean(cluster_silhouette_vals)
  print(f' Silhouette usando k = {cluster+1}')
  print(f'Score = {avg_score}')
```

><font color="006400"> **Método de la Silueta**
Haciendo un poco de teoria, el método de la silueta puede utilizarse para estudiar la distancia de separación entre los grupos resultantes. El gráfico de silueta muestra una medida de cuán cerca está cada punto en un clúster de los puntos en los clústeres vecinos y por lo tanto proporciona una forma de evaluar parámetros como el número de clústeres visualmente.

><font color="006400">Esta medida tiene un rango de -1 a 1. Donde 1 significa que los puntos están muy cerca de su propio clúster y lejos de otros clústeres, mientras que -1 indica que los puntos están cerca de los clústeres vecinos.
